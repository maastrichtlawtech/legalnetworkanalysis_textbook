```python
wk = nx.DiGraph()
```

```python
[wk.add_edge(x,y) for x, y in zip(df.country, df.beaten_by)]
```

```python
pos = nx.spring_layout(wk)
nx.draw_networkx_nodes(wk, pos=pos)
nx.draw_networkx_edges(wk, pos=pos)
nx.draw_networkx_labels(wk, pos=pos)
```

```python
W = nx.adjacency_matrix(wk).todense()
```

```python
W
```

```python
nodelist = list(wk.nodes)
```

```python
wkdf = pd.DataFrame(W, columns=nodelist, index=nodelist)
wkdf.to_csv("wkdf.csv")
```

```python
x = np.ones(8)/8
A = wkdf.values
M = A.T@A
for i in range(10):
    x = M@x
    xnorm = np.sum(x)
    x = x/xnorm
    print("authorities result --> \n", np.around(x,2))
    hubs = A@x
    hubsnorm = np.sum(hubs)
    hubs = hubs/hubsnorm
    print("hubs this iteration --> \n" , np.around(hubs, 2))
    sauth = A.T@hubs
    sauthnorm = np.sum(sauth)
    sauth = sauth/sauthnorm
    print("special authority output is \n", sauth)
```
